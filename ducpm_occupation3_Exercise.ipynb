### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.DataFrame(pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|', index_col = 'user_id'))
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users[users['gender'] == 'M'].groupby('occupation')['gender'].count()
```

```python
occo_all = users.groupby('occupation')['gender'].count()
```

```python
occo_m = users[users['gender'] == 'M'].groupby('occupation')['gender'].count()
```

```python
m_ratio = round((occo_m / occo_all)*100,2)
pd.DataFrame(m_ratio.sort_values(ascending = False))
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max', 'mean'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])['age'].agg(['min', 'max', 'mean'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
a = pd.DataFrame(round((users.groupby('occupation')['gender'].value_counts() / users.groupby('occupation')['gender'].count())*100, 2))
a
```
