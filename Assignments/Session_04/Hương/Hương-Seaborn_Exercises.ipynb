# Seaborn Exercises

## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## plot join distribution of 'fare' as x and 'age' as y using ``seaborn.joinplot``

```python
sns.jointplot(x = 'fare',y = 'age', data = titanic)
```

## plot distribution of 'fare' using ``seaborn.distplot``

```python
sns.distplot(titanic['fare'], bins = 30, color = 'red', kde = 0, hist = 1);
```

## plot distribution of 'age' with 3 difference 'class' using ``seaborn.boxplot``

```python
sns.boxplot(x = 'class', y = 'age', data = titanic,orient = "v", palette = ['blue','green','orange'])
```

## same as above, but this time let's use ``seaborn.swarmplot``

```python
sns.swarmplot(x = 'class', y = 'age', data = titanic, palette = ['green','orange','purple'])
```

## Use ``seaborn.countplot`` to show number of 'male' and 'female'

```python
sns.countplot(x = 'sex', data = titanic)
```

## Use ``pandas.corr`` to show correlation metrix for data fields

```python
titanic.corr()
```

## Then plot correlation matrix using ``seaborn.heatmap``

```python
sns.heatmap(data = titanic.corr())
plt.title("titanic.corr()");
```

## Use ``seaborn.FacetGrid`` mapped with ``plt.hist`` to plot histogram of ``age`` for 'male' & 'female'

```python
ans = sns.FacetGrid(data=titanic,col='sex')
ans.map(plt.hist,'age')
```
