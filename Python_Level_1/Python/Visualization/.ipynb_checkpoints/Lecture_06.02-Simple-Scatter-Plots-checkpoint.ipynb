# Simple Scatter Plots
![](https://seaborn.pydata.org/_images/seaborn-scatterplot-10.png)

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Scatter Plots with ``plt.plot``

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
#Notice the 3rd argument 'o'
plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

**These character codes can be used together with line and color codes to plot points along with a line connecting them:**

```python
plt.plot(x, y, '-ok');
```

**Additional keyword arguments to ``plt.plot`` specify a wide range of properties of the lines and markers:**

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

## Scatter Plots with ``plt.scatter``

```python
plt.scatter(x, y, marker='o');
```

**``plt.scatter``**: controll properties of each individual point (size, face color, edge color, etc.)  
**``alpha``** keyword to adjust the transparency level  
**``plt.colorbar``**: show color scale
- The color argument is automatically mapped to a color scale (shown here by the ``colorbar()`` command)
- The size argument is given in pixels.
- In this way, the color and size of points can be used to convey information in the visualization, in order to visualize multidimensional data.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

## Example: Iris data
![](https://miro.medium.com/max/2438/1*v34h7DKNcw74qtQxjpHUbA.png)

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
features
```

## 4 different dimensions of the data:
- The (x, y) location: sepal length and width
- The size of the point: petal width
- The color: species of flower.

```python
plt.scatter(features[0], features[1], alpha=0.3,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## ``plot`` Versus ``scatter``: A Note on Efficiency
### ``plt.plot`` should be preferred over ``plt.scatter`` for large datasets.
